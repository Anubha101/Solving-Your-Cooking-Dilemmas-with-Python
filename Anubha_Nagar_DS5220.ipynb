**Parameters:**<br>

**Example Input:** `1500, 2` <br>

```python
#importing the cookbook
import pandas as pd

cookbook = pd.read_csv("recipes.csv")

cookbook
```

```python

def total_calories(recipes= None):
    #if recipe is not given as an input then continue with the code
    if not recipes:
        #take input and ensure it is in camel case to match the cookbook format and stip any space before or after the word
        recipe_input = input("enter recipe names separated by a comma").title().strip()
        #make sure there is no digit
        recipe_input = ''.join([i for i in recipe_input if not i.isdigit()])
        #make sure input is numeric
        if recipe_input.isdigit():
            return "You have entered an incorrect recipe"
        #check if input is empty 
        if not recipe_input:
            return "No input was provided! Please check the input!"
        #split the input into different recipes
        recipes = [x.strip() for x in recipe_input.split(',')]
    #innitialize the variables for further calculation
    total_cal = 0
    total_serv =[]
    multiply = 0
    #iterate over for each recipe to idetify the servings and calorie per serving
    for i in recipes:
        serv = cookbook.loc[cookbook["Recipe"]== i ]["Servings"].unique()
        cal = cookbook.loc[cookbook["Recipe"]== i ]["Calories per Serving"].unique()
        if len(serv>0):
            a = serv#servings for each
        if len(cal>0):
            multiply += cal[0]*a#multiply the servings and calorie per servings
        
        else:
            print(i, "is not in cookbook! please enter valid recipe!")#if the recipe is not there in cookbook then print recipe to specify
    return (multiply)
#calling function
total_calories()
#checking for the following input:
#Bulgogi,Miso Soup,Paratha
```

```python
#need itertools for combinations
from itertools import combinations
import itertools
```

```python
def calories_recipes(calories= None, num_recipes = None):
    #user input for calories intake and number of recipes they want to make
    calories, num_recipes =map(int,input('\nWhat is the calory intake you want for your meal today and the number of recipe combinations eg 1500,2 if you want 1500 calories and 2 combination of recipes: \n').split(','))
    #empty list to store the possible combinations of recipes which have a colory value lesser than or equal to what the person inputted
    output_rec=[]
    #finding the unique recipes so we can understand what the different combinations of all possible calory values can be based on the number of combinations.
    rec_unique = cookbook.Recipe.unique()
    #finding all possible combinations for the particular number of recipes so that we can then we can check for which combination satisfies the calory limit
    combination_recipes = list(itertools.combinations(rec_unique,num_recipes))
    #iterate over all combinations of recipes
    for i in combination_recipes:
        #find the total calory value of both
        tot_cal = total_calories(list(i))
        #check if sum of calory value is less than or equal to the calory limit entered by the user
        if tot_cal <= calories:
            output_rec.append(i)#if true, append into the list
    if len(output_rec) == 0:#checkig for the case where no recipe combinations in the cookbook match the calorie value entered
        print("Nothing matches the calory value you entered! Please think of another calory intake!")
    #print in the same line
    print(output_rec,end = '\r')
```

```python
#same code as basic questions
import pandas as pd
def view_ingredients(cookbook, recipe):
    indx = []
    for i in range(len(cookbook.Recipe)):
        if recipe == cookbook.Recipe[i]:  
            indx.append(i)
    df = pd.DataFrame()
    for indexes in indx:
        df = df.append(cookbook.iloc[indexes])
        df.dropna()
        print_fn(df.drop_duplicates(subset=None, keep="first", inplace=False))
```

```python
#printing the output in the correct format
def print_fn(df):
    #name of recipe
    name_str = ' '.join([str(k) for k in df.Recipe.unique()])
    print("Name:" + " " + name_str)
    #one blank line
    print(" ")
    #style of recipe
    sty_str = ' '.join([str(k) for k in df.Style.unique()])
    print("Style: "+ sty_str)
   
    print("Recipe: ")
    #quantity and ingredients of recipe
    ans = df[["Quantity", "Ingredient"]]
    for index, row in df.iterrows():
        print(row['Quantity'] + ' ' + row['Ingredient'])
    #servings
    print("Servings: ")
    ser_str = ' '.join([str(k) for k in df.Servings.unique()])
    print(ser_str)
    #calories per serving
    print("Calories per Serving: ")
    print(df["Calories per Serving"].iloc[0])
```
